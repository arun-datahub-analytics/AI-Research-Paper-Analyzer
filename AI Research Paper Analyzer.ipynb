```python
# 2. Extract Text from Research Paper
```

```python
import fitz

pdf = fitz.open("Paper.pdf")

text = ""

for page in pdf:
    text += page.get_text()

print("Total Characters:", len(text))
print("\n")
print(text[:3000])
```

```python
# 3. Text Cleaning and Preprocessing
```

```python
import re

clean_text = re.sub(r'\s+', ' ', text)
clean_text = clean_text.strip()

print("Cleaned Characters:", len(clean_text))
print(clean_text[:2000])
```

```python
# 4. Gemini LLM Configuration
```

```python
import google.generativeai as genai

genai.configure(api_key="YOUR_APP_KEY")

model = genai.GenerativeModel("gemini-2.5-flash")

print("Gemini Connected Successfully")
```

```python
# 5. AI-Powered Research Paper Summarization
```

```python
prompt = f"""
Analyze this research paper and provide:

1. Paper Title
2. Research Problem
3. Key Contributions
4. Methodology
5. Important Findings
6. Limitations
7. Future Research Directions
8. Simple Explanation for Beginners

Research Paper:

{clean_text[:15000]}
"""

response = model.generate_content(prompt)

print(response.text)
```

```python
# 6. Research Paper Keyword Extraction
```

```python
prompt = f"""
Extract the top 20 important technical keywords from this research paper.

Research Paper:
{clean_text[:15000]}
"""

response = model.generate_content(prompt)

print(response.text)
```

```python
# 7. AI-Based Research Gap Identification
```

```python
prompt = f"""
You are an AI Research Scientist.

Analyze this research paper and identify:

1. Research gaps
2. Limitations of the proposed method
3. Unsolved problems
4. Future improvement opportunities
5. Potential research ideas based on this paper

Research Paper:

{clean_text[:15000]}
"""

response = model.generate_content(prompt)

print(response.text)
```

```python
# 8. Literature Review Generator
```

```python
prompt = f"""
You are an AI Research Scientist.

Analyse the following research paper and generate a detailed literature review.

Include:

1. Introduction
2. Objective of the Paper
3. Methodology
4. Key Findings
5. Advantages
6. Limitations
7. Future Scope
8. Conclusion

Research Paper:

{clean_text[:15000]}
"""

response = model.generate_content(prompt)

print(response.text)
```

```python
# 9. Research Paper Question Answering System
```

```python
question = input("Ask a question about the research paper: ")

prompt = f"""
You are an AI Research Scientist.

Answer the question using only the information available in the research paper.

Research Paper:

{clean_text[:15000]}

Question:

{question}

Provide:
1. Direct Answer
2. Explanation
3. Supporting Evidence from the Paper
"""

response = model.generate_content(prompt)

print(response.text)
```

```python
# 10. Export Research Analysis Report
```

```python
with open("Research_Report.txt", "w", encoding="utf-8") as f:
    f.write(response.text)

print("Research Report Saved Successfully")
```
